# Music & Motivation Survey — Starter EDA

This notebook loads the cleaned/anonymized survey data, runs basic checks, and makes a few starter plots.

**Note:** `data/raw/` is gitignored because it may contain identifying information.


```
import pandas as pd
import matplotlib.pyplot as plt

DATA_PATH = '../data/processed/survey_clean_anonymized.csv'
df = pd.read_csv(DATA_PATH, parse_dates=['Timestamp'])
df.head()
```

```
df.shape, df.columns.tolist()
```

```
# Missingness
(df.isna().mean().sort_values(ascending=False).head(15))
```

```
# Example: distribution of introversion/extroversion rating
col = [c for c in df.columns if 'how introverted' in c.lower()][0]
df[col].value_counts().sort_index().plot(kind='bar')
plt.title('Introversion/Extroversion self-rating (1–5)')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```
# Example: compare conversation likelihood with no music vs loud music
no_music = [c for c in df.columns if 'no music' in c.lower()][0]
loud_music = [c for c in df.columns if 'loud music' in c.lower()][0]

plot_df = pd.DataFrame({
    'No music': df[no_music],
    'Loud music': df[loud_music]
})
plot_df.mean().plot(kind='bar')
plt.title('Mean likelihood of starting a conversation (1–5)')
plt.ylabel('Mean rating')
plt.show()

plot_df['delta_loud_minus_no'] = plot_df['Loud music'] - plot_df['No music']
plot_df['delta_loud_minus_no'].describe()
```
